# Exploratory Data Analysis

### Class


```python
ax = sns.countplot(x = 'Class',data = df)
sns.set(rc={'figure.figsize':(5,11)})

for bars in ax.containers:
    ax.bar_label(bars)
```

*From above graphs we can see which section have max and min no of students of class 10th*

```python
subject_avg = df[['Math', 'Science', 'English','Social Studies']].mean()
print(subject_avg)
```

```python
subject_avg.plot(kind = 'bar', color = 'lightpink')
plt.title('Average marks per subject')
plt.xlabel('Subjects')
plt.ylabel('average score')
plt.tight_layout()
sns.set(rc={'figure.figsize':(10,10)})
plt.show()
```

***From the above bar graph we can analyze the average marks scored by all students in each subject***

```python
# Student name  vs Average marks
avg = df.groupby(['Student Name'], as_index=False)['Average Score'].sum().sort_values(by='Average Score', ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x = 'Student Name',y= 'Average Score' ,data = avg,color = 'lightgreen')
```

***from above analysis we can find the topper student by average score***

```python
avg = df.groupby(['Student Name'], as_index=False)['Attendance (%)'].sum().sort_values(by='Attendance (%)', ascending=False).head(20)
sns.set(rc={'figure.figsize':(25,7)})
sns.barplot(x = 'Student Name',y= 'Attendance (%)' ,data = avg , color = 'indigo')
```

***frome above analysis we can see that which student is present in most classes***

```python
plt.bar(df['Attendance (%)'],df['Average Score'],color = 'red')
plt.xlabel('Attendance (%)')
plt.ylabel('Average score')
plt.grid
plt.show
```

***from above bar we can compare that students which have high attendance rate have high average score***

```python
sns.set_context("poster")
ax = sns.barplot(data = df, x = 'Grade' , y = 'Attendance (%)', hue = 'Class')
sns.set_context("poster")
sns.set(rc={'figure.figsize':(20,20)})


```

***using seaborn library we can see that grade vs attendance relationship***

```python
df['performance category'] = df['Average Score'].apply(lambda x : 'High performer' if x>= 75 else 'Low performer')
sns.countplot(data = df , x = 'performance category' , hue = 'Class')
sns.set(rc={'figure.figsize':(5,10)})
plt.title('Count of High Performer vs Low Performer')
```

```python
subjects = ['Math','English','Social Studies','Science']
df_small = df.head(10)
df_melted = df_small.melt(id_vars = ['Student Name'],value_vars = subjects,var_name = 'Subjects',value_name = 'Marks')
sns.barplot(data = df_melted , x = 'Student Name' , y = 'Marks' , hue = 'Subjects')
sns.set(rc={'figure.figsize':(25,10)})

```

***from above grouped graph we can compare the scores over different subjects of a particular student. Thus helps in determing the weak points of the student***

```python
columns = ['Attendance (%)','Average Score', 'Math' , 'Science' , 'English']
sns.heatmap(data = df[columns].corr() , annot = True , cmap = 'Blues')
sns.set(rc={'figure.figsize':(20,10)})
```

***using heatmaps we can see that correlation beteween over different columns***

```python
ax = sns.countplot(data = df , x = 'Grade' , hue = 'Class')
for bars in ax.containers:
    ax.bar_label(bars)
```
